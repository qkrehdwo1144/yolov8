**라이브러리 설치**

```python
!pip install ultralytics
import torch
from PIL import Image
import requests
from io import BytesIO
```

**다운로드/로드 이미지**

```python
image_url = 'https://www.ibabynews.com/news/photo/202009/88976_38171_1626.jpg'
response = requests.get(image_url)
img = Image.open(BytesIO(response.content))
```

**Yolo모델 불러오기**

```python
from ultralytics import YOLO
```

```python
# Load a pre-trained YOLOv8 model (YOLOv8s is a small version, but you can try larger versions like YOLOv8m, YOLOv8l)
model = YOLO('yolov8s.pt')
```

```python
# Step 4: Run YOLOv8 on the image
results = model(img)
```

```python
# Step 5: Show results
# Extract and display the image with bounding boxes
results[0].plot()  # This should plot the image with detections
```

```python
# If you want to show the image using PIL
img_with_boxes = results[0].plot()  # Returns the image with boxes
Image.fromarray(img_with_boxes)  # Convert back to PIL image and display
```
